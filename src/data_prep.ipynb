```python
import pandas as pd 
import pyodbc
```

```python
# Conectando com o banco de dados sql server
conn = pyodbc.connect('Driver={SQL Server};'
                      'Server=BEAUTYBALL\SQLEXPRESS;'
                      'Database=decision_db;'
                    'Trusted_Connection=yes;')
cursor = conn.cursor()
```

```python
# Lendo dados de uma tabela
dados = pd.read_sql_query('SELECT * FROM decision_db.dbo.feature_store', conn)
dados.head()
```

```python
# Verificando shape do dataframe
dados.shape
```

```python
# Verificando quantidade de valores nulos por coluna 
dados.isnull().sum()
```

Filtrando apenas para que retorne curriculos prospectados e que não sejam iguais a vazio (não faz sentido tê-los na base)

```python
# Verificando a quantidade de nulos em situacao_candidato
dados['situacao_candidato'].isnull().sum()
```

```python
# Ordenando a quantidade de nulos por situacao_candidato
dados['situacao_candidato'].value_counts()
```

```python
# Filtrando apenas os que geraram prospects
dados_prospectados = dados[dados['situacao_candidato'] == 'Prospect']
dados_prospectados
```

```python
# Verificando se exisem curriculos vazios (isto é, igual a '')
dados_prospectados[dados_prospectados['curriculo'] == '']
```

```python
# Limpando curriculos vazios
dados_prospectados = dados_prospectados[dados_prospectados['curriculo'] != '']
```

```python
dados_prospectados.reset_index(drop=True, inplace=True)
```

```python
dados_prospectados.shape
```

```python
# Salvando a base de dados em um csv em tc5_fiap_datathon/data
dados_prospectados.to_csv('../data/dados_prospectados.csv', index=False)
```
